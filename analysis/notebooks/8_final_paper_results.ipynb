## Stage 1

```python
base_path = "../data/stage1/"
directories = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]
directories
```

```python
all_data = []

# Iterate over each directory
for directory in directories:
    # Construct the path to participant_info.json
    file_path = os.path.join(base_path, directory, 'participant_info.json')
    
    # Check if the file exists
    if os.path.exists(file_path):
        # Open and read the json file
        with open(file_path, 'r') as file:
            for line in file:
                # Parse each line as JSON and append to the list
                all_data.append(json.loads(line))

# Convert the list of JSON objects into a DataFrame
participant_info = pd.DataFrame(all_data)
```

```python
participant_info = participant_info[participant_info['prolific_id']!="65e6fb8c62b34b6ef5249890"]
participant_info['timestamp'] = participant_info['timestamp'].apply(lambda x: x['$date'])
participant_info['timestamp'] = pd.to_datetime(participant_info['timestamp'])
participant_info['_id'] = participant_info['_id'].apply(lambda x: x['$oid'])
participant_info = participant_info.drop_duplicates()
participant_info.head(1)
```

```python
final_document_list = pd.read_csv('../data/buckets_combined/master_documents_with_pid.csv')
final_document_list.head(1)
```

```python
participant_info['prolific_id'] = participant_info['prolific_id'].str.replace('ABCDEF', '66617548416d40fa86a254a0')
```
