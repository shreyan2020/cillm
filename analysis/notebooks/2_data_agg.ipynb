# Documents

```python
all_data = []

# Iterate over each directory
for directory in directories:
    # Construct the path to participant_info.json
    file_path = os.path.join(base_path, directory, 'document.json')
    
    # Check if the file exists
    if os.path.exists(file_path):
        # Open and read the json file
        with open(file_path, 'r') as file:
            for line in file:
                # Parse each line as JSON and append to the list
                all_data.append(json.loads(line))

# Convert the list of JSON objects into a DataFrame
document = pd.DataFrame(all_data)
document
```

```python
# document['timestamp'] = document['timestamp'].apply(lambda x: x['$date'])
# document['timestamp'] = pd.to_datetime(document['timestamp'])
document['_id'] = document['_id'].apply(lambda x: x['$oid'])
document = document[document['prolific_id']!='65e6fb8c62b34b6ef5249890']
```

```python
def extract_task_and_study_id(task_name):
    parts = task_name.split('_')
    task_id = '_'.join(parts[:-1])  # All parts except the last one
    study_id = parts[-1]  # The last part
    return task_id, study_id

# Apply the function to the 'task_name' column and store the results in two new columns
document['task_id'], document['study_id'] = zip(*document['task_id'].apply(extract_task_and_study_id))
```

```python
def extract_text_and_word_count(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    plain_text = soup.get_text()
    word_count = len(plain_text.split())
    return plain_text, word_count

document['plain_text'], document['word_count'] = zip(*document['content'].apply(extract_text_and_word_count))
```

```python
def is_valid_submission(group):
    # Check if both 'main_task_1' and 'main_task_2' are in the 'task_id' column for this group
    return all(task in group['task_id'].values for task in ['main_task_1', 'main_task_2'])

document[document['study_id']!='66c516b1d61c7b572205f713'].groupby(['study_id','prolific_id']).filter(is_valid_submission)['study_id'].value_counts()
```
